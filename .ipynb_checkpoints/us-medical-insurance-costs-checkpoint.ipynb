# U.S. Medical Insurance Costs

```python
# import Python's built-in csv module
import csv
```

```python
# create an empty list where the rows will be stored
data = []

# open the CSV file, read each row as a dictionary and add it to the data list
with open('insurance.csv') as insurance:
    reader = csv.DictReader(insurance)
    for row in reader:
        data.append(row)
```

```python
# print column names
print("Column names:", reader.fieldnames)

# print first 5 rows
for row in data[:5]:
    print(row)
```

```python
# compare bmi of smokers vs nonsmokers
bmi_smokers = []
bmi_nonsmokers = []

for row in data:
    bmi = float(row['bmi'])
    if row['smoker'] == 'yes':
        bmi_smokers.append(bmi)
    else:
        bmi_nonsmokers.append(bmi)

avg_bmi_smokers = sum(bmi_smokers)/len(bmi_smokers)
avg_bmi_nonsmokers = sum(bmi_nonsmokers)/len(bmi_nonsmokers)

print("Average BMI of smokers:", round(avg_bmi_smokers, 2))
print("Average BMI of non-smokers:", round(avg_bmi_nonsmokers, 2))
print('Smoking status does not appear to affect BMI in this dataset — smokers and non-smokers have almost identical average BMI')
```

```python
# find the proportion of smokers/nonsmokers in different regions

# count smokers/non-smokers per region
counts = {}
for row in data:
    region = row['region']
    smoker = row['smoker']
    
    if region not in counts:
        counts[region] = {"yes": 0, "no": 0}
    
    counts[region][smoker] += 1

# calculate proportions
proportions = {}
for region, vals in counts.items():
    total = vals["yes"] + vals["no"]
    proportions[region] = {
        "yes": vals["yes"] / total,
        "no": vals["no"] / total
    }

print("Counts per region:")
for region, vals in counts.items():
    print(region, vals)

print("\nProportions per region:")
for region, vals in proportions.items():
    print(region, vals)

print('The southeast region has a noticeably higher share of smokers compared to other regions, while the southwest and northwest have the lowest')
```

```python
# how many kids usually have smokers/nonsmokers
children_stats = {"yes": {"total_children": 0, "count": 0},
                  "no": {"total_children": 0, "count": 0}}

for row in data:
    smoker = row["smoker"]
    children = int(row["children"])
    
    children_stats[smoker]["total_children"] += children
    children_stats[smoker]["count"] += 1

# Compute averages
averages = {}
for smoker, vals in children_stats.items():
    if vals["count"] > 0:
        averages[smoker] = vals["total_children"] / vals["count"]
    else:
        averages[smoker] = 0

print("Average children of smokers:", averages["yes"])
print("Average children of non-smokers:", averages["no"])
print('Smoking status does not appear to be associated with the number of children people have in this dataset. Both smokers and non-smokers usually have about 1 child on average')
```

```python
# how BMI for women changes depending on the number of children they have
# how many women are in each group of children
bmi_by_children = {}

for row in data:
    if row["sex"] == "female":   # only women
        children = int(row["children"])
        bmi = float(row["bmi"])
        
        if children not in bmi_by_children:
            bmi_by_children[children] = {"total_bmi": 0, "count": 0}
        
        bmi_by_children[children]["total_bmi"] += bmi
        bmi_by_children[children]["count"] += 1

# Compute averages
for children, vals in sorted(bmi_by_children.items()):
    avg_bmi = vals["total_bmi"] / vals["count"]
    print(f"Children = {children}: Average BMI = {avg_bmi:.2f}, Count = {vals['count']}")

print("Women’s BMI in this dataset is quite stable regardless of number of children, with only a slight increase among those with 4 children")
print("Number of children doesn’t strongly affect women’s BMI in this dataset. The apparent increase at 4 kids is based on very few cases, so it’s not statistically reliable")
```

```python
#calculate average charges for smokers/nonsmokers
charges_stats = {"yes": {"total": 0, "count": 0}, "no": {"total": 0, "count": 0}}

for row in data:
    smoker = row["smoker"]
    charge = float(row["charges"])
    
    charges_stats[smoker]["total"] += charge
    charges_stats[smoker]["count"] += 1

for smoker, vals in charges_stats.items():
    avg = vals["total"] / vals["count"]
    print(f"Average charges for {'smokers' if smoker=='yes' else 'non-smokers'}: {avg:.2f}")

print('Smoking status is the key driver of insurance charges: smokers incur significantly higher costs than non-smokers')
```

```python
# total charges and counts per region and smoker status
charges_by_region = {}

for row in data:
    region = row["region"]
    smoker = row["smoker"]
    charge = float(row["charges"])
    
    if region not in charges_by_region:
        charges_by_region[region] = {"yes": {"total": 0, "count": 0},
                                     "no": {"total": 0, "count": 0}}
    
    charges_by_region[region][smoker]["total"] += charge
    charges_by_region[region][smoker]["count"] += 1

# Compute averages
for region, vals in charges_by_region.items():
    avg_yes = vals["yes"]["total"] / vals["yes"]["count"] if vals["yes"]["count"] > 0 else 0
    avg_no = vals["no"]["total"] / vals["no"]["count"] if vals["no"]["count"] > 0 else 0
    print(f"Region: {region}")
    print(f"  Average charges for smokers: {avg_yes:.2f}")
    print(f"  Average charges for non-smokers: {avg_no:.2f}\n")

print('Smoking status is the strongest factor affecting insurance charges')
print('Regional differences exist but are much smaller, especially for non-smokers')
print('Southeast smokers are the highest payers, Northeast non-smokers the lowest')
```
